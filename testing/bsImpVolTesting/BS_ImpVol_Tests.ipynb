```python

BS_analyticalPrice = np.empty_like(arr_Option_MktPrices)

df_BSImpVol = df_OptionPrices_market_r[[x for x in df_OptionPrices_market_r.columns if x not in cols_strike]]
df_OptionPrices_market_r.loc[:,"rfr"] = [r for i in range(0,len(df_OptionPrices_market_r))]

for i in range(0,len(cols_strike)):

    col = cols_strike[i]
    df_BSImpVol[col] = df_OptionPrices_market_r.apply(lambda x: EuroOption.getBS_ImpVol(x[col],
                                                                                        x['Underlying Value'],
                                                                                        int(col),
                                                                                        x['rfr'],
                                                                                        x['DaystoExpiry']/360,
                                                                                        "call")
                                                      ,axis=1)
    S0 = np.array(df_OptionPrices_market_r.loc[:,'Underlying Value'])
    strike = int(col)
    rfr = np.array(df_OptionPrices_market_r.loc[:,'rfr'])
    vol = np.array(df_BSImpVol.loc[:,col])
    maturity = np.array(df_OptionPrices_market_r.loc[:,'DaystoExpiry'])/360
    optionType="call"
    BS_analyticalPrice[:,i] = EuroOption.price_BS_analytical(S0,strike,rfr,vol,maturity,optionType="call")


    

diff = arr_Option_MktPrices - BS_analyticalPrice


df_diff = pd.DataFrame(data=diff,
                     columns=cols_strike,
                     index=timeToMat)

df_BSAnalyticalPrices = pd.DataFrame(data=BS_analyticalPrice,
                     columns=cols_strike,
                     index=timeToMat)

BS_IVs = df_BSImpVol[cols_strike].to_numpy()

path_surface = outputPath + "BSImpVol_Surface"
path_diff = outputPath + "MarketVsBS_Diff_Surface"
path_priceSurface = outputPath + "Analytical_Price_Surface"

EuroOption.plot_surface(BS_IVs, K, timeToMat, x_label="Strikes", y_label="Time to Maturity", z_label="BS Imp Vol",
             title="BS ImpVol Surface", save=True, filePath=path_surface+".png",show=True)
EuroOption.plot_surface(diff, K, timeToMat, x_label="Strikes", y_label="Time to Maturity", z_label="Price Diff",
             title="Price Diff Surface", save=True, filePath=path_diff+".png",show=True)


writer = pd.ExcelWriter(path_surface+".xlsx", engine='xlsxwriter')
workbook = writer.book
worksheet1 = workbook.add_worksheet('BS_ImpVolSurface')
worksheet2= workbook.add_worksheet('diff_MarketPriceVsBSPrice')
worksheet3 = workbook.add_worksheet('BS_analyticalPrice')
writer.sheets['BS_ImpVolSurface'] = worksheet1
writer.sheets['diff_MarketPriceVsBSPrice'] = worksheet2
writer.sheets['BS_analyticalPrice'] = worksheet3

df_BSImpVol.to_excel(writer, sheet_name='BS_ImpVolSurface', startrow=0, startcol=0)
df_diff.to_excel(writer, sheet_name='diff_MarketPriceVsBSPrice', startrow=0, startcol=0)
df_BSAnalyticalPrices.to_excel(writer, sheet_name='BS_analyticalPrice', startrow=0, startcol=0)
writer.save()
```
